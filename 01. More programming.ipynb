Dans cet exercice, vous allez écrire une fonction Python qui utilise **la loi de Darcy** pour calculer le débit à travers un milieu poreux.

## Description du problème

La loi de Darcy stipule que le débit ($Q$) à travers un milieu poreux est donné par :

$$ 
Q = k \times A \times \left(\frac{\Delta h}{L}\right)
$$ 

Où :
- **$k$** est la conductivité hydraulique (en mètres par seconde, $m/s$).
- **$A$** est l'aire de la section transversale (en mètres carrés, $m^2$).
- **$\Delta h$** est la différence de charge hydraulique (en mètres, $m$).
- **$L$** est la distance sur laquelle s'exerce la différence de charge hydraulique (en mètres, $m$).

**Votre tâche :**

1. **Écrivez une fonction** `darcy_flow(k, A, delta_h, L)` qui :
   - Accepte quatre paramètres : `k`, `A`, `delta_h` et `L`.
   - Calcule le gradient hydraulique ($\frac{\Delta h}{L}$).
   - Renvoie le débit ($Q$) en utilisant la loi de Darcy.
   
2. **Testez votre fonction** en calculant les débits pour une plage de différences de charge hydraulique (`delta_h`), en gardant `k`, `A` et `L` constants.

3. **Tracez les résultats :** Créez un graphique du débit ($Q$) en fonction de la différence de charge hydraulique ($\Delta h$).

## Conseils utiles

- Utilisez le mot-clé `def` pour définir votre fonction.
- Le gradient hydraulique se calcule comme `delta_h / L`.
- Vous pouvez utiliser la fonction `linspace` de la bibliothèque `numpy` pour créer un tableau de valeurs de `delta_h`.

C'est parti !


```python
# Complétez la fonction ici
def darcy_flow(k, A, delta_h, L):  # Finalisez en indiquant TOUS LES PARAMÈTRES
    """
    Calcule le débit Q en utilisant la loi de Darcy.
    
    Paramètres :
        k : Conductivité hydraulique (m/s)
        A : Aire de la section transversale (m^2)
        delta_h : Différence de charge hydraulique (m)
        L : Distance sur laquelle s'exerce la différence de charge (m)
    
    Renvoie :
        Q : Débit Q (m^3/s)
    """
    
    Q =  k * A * (delta_h / L)  # INSÉRER L'EXPRESSION ICI
    
    return Q

# Vous pouvez tester votre fonction avec un exemple simple :
print("Sortie de test de la fonction:", darcy_flow(1e-5, 10, 5, 100))
```

## Étape 2 : Calculer les débits pour une plage de différences de charge hydraulique

Utilisons maintenant votre fonction pour calculer le débit \( Q \) pour une plage de différences de charge hydraulique (`delta_h`), tout en gardant les autres paramètres constants.

Par exemple, supposons :
- $ k = 1 \times 10^{-5} $ m/s,
- $ A = 10 $ m$^2$,
- $ L = 100 $ m.

Nous ferons varier `delta_h` de 0.1 m à 10 m.


```python
import numpy as np

# Paramètres constants donnés
k = 1e-5  # Conductivité hydraulique en m/s
A = 10    # Aire de la section transversale en m^2
L = 100   # Longueur en m

# Créer un tableau de valeurs de delta_h allant de 0.1 m à 10 m (en évitant 0 pour prévenir tout problème de division)
delta_h_values = np.linspace(0.1, 10, 50)

# Calculer Q pour chaque valeur de delta_h en utilisant votre fonction darcy_flow
Q_values = []
for dh in delta_h_values:
    Q_values.append(darcy_flow(k, A, dh, L))

# Afficher quelques valeurs calculées pour vérifier vos résultats
print("Exemple de valeurs de Q calculées:", Q_values[:5])
```

## Étape 3 : Tracer les résultats

Enfin, tracez le débit $ Q $ en fonction de la différence de charge hydraulique $ \Delta h $ en utilisant `matplotlib`.

Ce graphique aidera à visualiser comment le débit augmente avec l'augmentation de la différence de charge hydraulique.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(delta_h_values, Q_values, marker='o', linestyle='-', color='steelblue')
plt.xlabel('Différence de charge hydraulique (Δh) [m]')
plt.ylabel('Débit Q [m³/s]')
plt.title("Loi de Darcy : débit vs. différence de charge hydraulique")
plt.grid(True)
plt.show()
```
