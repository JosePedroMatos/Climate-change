# Visualiser des données
### Vue d'ensemble

Les données CORDEX correnpondent à des matrices 3D (date, latitude et longitude).  
Avec le module meteoraster, it est relativement facile de les travailler.

Ici, nous allons explorer une projéction CORDEX couvrant toute l'Afrique.

### Charger les données avec MeteoRaster
Pour le faire, il est nécéssaire de fournir la position (rélative) du fichier.  
Dans cette exemple, il est dans le dossier `data`du projet.

```python
#Importer les modules nécéssaires
from pathlib import Path
from meteoraster import MeteoRaster
from matplotlib import pyplot as plt

#Charger les données CORDEX
data_path = Path(r'data/tas_AFR-44_ICHEC-EC-EARTH_historical_r12i1p1_KNMI-RACMO22T_v1_mon_19500101-20051231.mr')
data = MeteoRaster.load(data_path)
```

### Préparer la figure
En utilisant les fonctions `create_plot`et `plot_mean`, il est possible d'obtenir de générer des figures.  
`cmap` designe les couleurs utilisées. Le choix du "color map" peut être faite avec l'aide de cette référence:  
https://matplotlib.org/stable/gallery/color/colormap_reference.html

```python
cmap = 'magma'

ax = data.create_plot(central_longitude=20)
fig = data.plot_mean(ax=ax, coastline=True, borders=True, colorbar=True,
                  colorbar_label=f'[{data.units}]', cmap=cmap, central_longitude=20, central_latitude=0)
#plt.gcf().savefig(plt_folder / cropped_file_path.name.replace('.mr', '.png'), dpi=300)
```
